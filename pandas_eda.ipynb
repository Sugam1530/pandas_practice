# **Finding the number of apps present in each category and showing a bar chart to compare the number**

```python
category_wise_count = df_copy.groupby('Category')['App'].count()
```

```python
top_5_category_wise_count = category_wise_count.sort_values().nlargest(5)
```

```python
# Showing the chart using matplotlib where comparing the number of apps for different categories

top_5_category_wise_count.plot(kind='bar', figsize=(8, 4), color='skyblue')
plt.title('Number of Apps in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **Finding the highest rated app category among all**

```python
category_avg_ratings = df_copy.groupby('Category')['Rating'].mean()
```

```python
category_avg_ratings_sorted = category_avg_ratings.sort_values().nlargest(5)

# Create a horizontal bar chart
category_avg_ratings_sorted.plot(kind='barh', figsize=(8, 4), color='orange')

# Set chart title and labels
plt.title('Average Ratings by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')

# Show the plot
plt.show()
```

# **Finding the highest installed app category**

```python
df_copy['Installs'] = df_copy['Installs'].str.split('+')

# Extracting the numerical part and converting into integer
df_copy['Installs'] = df_copy['Installs'].str[0].str.replace(',', '').astype(int)
```

```python
df_copy.head()
```

```python
# filtering top 5 Installed categories of APP
category_total_installs = df_copy.groupby('Category')['Installs'].sum()
top_5_category_total_installs = category_total_installs.sort_values().nlargest(5)
```

```python
# Showing the pie chart

plt.figure(figsize = (10,4))
plt.pie(top_5_category_total_installs, labels = top_5_category_total_installs.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of total installs accross all categories')

plt.show()
```

# **Finding the number of apps present in each content Rating and showing a line chart to compare the number**

```python
df_copy['Content Rating'].value_counts()
```

```python
# showing the data through a line chart

# Create a horizontal bar chart
df_copy['Content Rating'].value_counts().plot(kind='line', figsize=(8, 4), color='blue')

# Set chart title and labels
plt.title('Different Content Types')
plt.xlabel('Contents')
plt.ylabel('Numbers')

# Show the plot
plt.show()
```

# **Showing the correlation between apps rating and installs with scatter chart and heat map**

```python
 df_copy.head()
```
